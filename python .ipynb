```python
def nameage(name,age,location,village):
    print("my name is",name)
    print("my age is",age)
    print("my current location is",location)
    print("my village is",village)

print("[*case:1*]")
nameage("ashish",18,"delhi","lucknow")
print("[case:2]")
nameage("pankaj",18,"delhi","lucknow")
print("[case:3]")
nameage("sonu",18,"delhi","rajsthan")
print("[case:4]")
nameage("rahul",26,"delhi","patna")
```
